# Articles

```python
url = 'https://www.henryakissinger.com/articles/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
links = [link['href'] for link in soup.find_all('a', {'class':'archive-item__permalink'})]
```

### Scraping all links

```python
all_articles_links = []
for i in range(1,11):
    url = f'https://www.henryakissinger.com/articles/page/{i}/'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    
    link_tags = soup.find_all('a', {'class':'archive-item__permalink'})
    for link_tag in link_tags:
        all_articles_links.append(link_tag['href'])
```

```python
len(all_articles_links)
for article_link in all_articles_links:
    print(article_link)
```

### Scraping content from links

```python
url = 'https://www.henryakissinger.com/articles/how-to-resolve-the-north-korea-crisis/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'lxml')
```

```python
soup.title
```

```python
title = (soup.title.text).split('|')[0]
title
```

```python
journal = soup.find('div', {'class':'post-publisher'}).text
journal
```

```python
publish_date = soup.find('div', {'class':'post-date'}).text
publish_date
```

```python
article_content = soup.find('div', {'class':'blog-content'}).text
article_content
```

```python
def get_article_detail(url):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    
    article_detail = dict()
    
    try:
        article_detail['title'] = (soup.title.text).split('|')[0]
    except:
        article_detail['title'] = None
    try:
        article_detail['journal'] = soup.find('div', {'class':'post-publisher'}).text
    except:
        article_detail['journal'] = None
    try:
        article_detail['publish_date'] = soup.find('div', {'class':'post-date'}).text
    except:
        article_detail['publish_date'] = None
    try:
        article_detail['article_content'] = soup.find('div', {'class':'blog-content'}).text
    except:
        article_detail['article_content'] = None
        
    return article_detail
        
```

```python
all_article_details = []
for article_link in all_articles_links:
    print(article_link)
    all_article_details.append(get_article_detail(article_link))
```

```python
all_article_details_df = pd.DataFrame(all_article_details)
all_article_details_df
```

```python
all_article_details_df.to_csv('Henry A.kissinger website data/articles.csv')
```

```python
def get_interview_detail(url):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    
    interview_detail = dict()
    
    try:
        interview_detail['title'] = (soup.title.text).split('|')[0]
    except:
        interview_detail['title'] = None
    try:
        interview_detail['publisher'] = soup.find('div', {'class':'post-publisher'}).text
    except:
        interview_detail['publisher'] = None
    try:
        interview_detail['publish_date'] = soup.find('div', {'class':'post-date'}).text
    except:
        interview_detail['publish_date'] = None
    try:
        interview_detail['interview_detail_content'] = soup.find('div', {'class':'blog-content'}).text
    except:
        interview_detail['interview_detail_content'] = None
        
    return interview_detail
```
